```python
print("Загадай целое число от -1000 до 1000, а я попробую его угадать.")

min = -1000
max = 1000

while min <= max:
    mid = (min + max) // 2
    print(f"Твоё число — {mid}?")

    answer = input("Введи: '=' если да, '+' если больше, '-' если меньше: ")

    if answer == '=':
        print("Я угадал число!")
        break
    elif answer == '+':
        min = mid + 1
    elif answer == '-':
        max = mid - 1
    else:
        print("Ало, меня слышно ты не то ввёл. Используй только '=', '+' или '-'.")
```
