```python
cache = {0: 0, 1: 1, 2: 1, 3: 2, 4: 3}

def fibonacci(n):
    if n in cache:
        return cache[n]
    result = fibonacci(n - 1) + fibonacci(n - 2)
    cache[n] = result
    return result

n = int(input("Введите номер числа Фибоначчи: "))
answer = fibonacci(n)
print(f'{n}-е число Фибоначчи: {answer}')
```
