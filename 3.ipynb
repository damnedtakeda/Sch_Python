```python
n = int(input("Сколько строк треугольника Паскаля вывести? "))

triangle = [[1]]
current_row = [1]

for _ in range(n - 1):
    temp_row = [0] + current_row + [0]
    next_row = []
    
    for i in range(len(temp_row) - 1):
        next_row.append(temp_row[i] + temp_row[i + 1])
    
    current_row = next_row
    triangle.append(current_row)

spaces = n - 1
for row in triangle:
    row_str = ' '.join(str(x) for x in row)
    print(' ' * spaces + row_str)
    spaces -= 1
```
