```python
from rich import print as rich_print

name = input("Введите ваше имя: ")

def change_name(name):
    resu = ""
    for i in range(len(name)):
        if i % 2 == 1:  
            resu += name[i].upper()
        else:
            resu += name[i].lower()
    return resu

def greet(name):
    rich_print(f"Привет [blue]{name}[/blue]")

changed = change_name(name)
greet(changed)

```
